### Agrégation des chiffres par région

```python
datareg = dataREG.groupby(['REG']).sum().reset_index()
datareg.shape
```

### Agrégation en 3 classes d'âges

```python
groupby_dict = {"P16_POP":"P16_POP",
                "P16_POP0014":"P16_POP0029",
                "P16_POP1529":"P16_POP0029", 
                "P16_POP3044":"P16_POP3059",
                "P16_POP4559":"P16_POP3059",
                "P16_POP6074":"P16_POP60P",
                "P16_POP75P":"P16_POP60P"
                } 
  
datareg = datareg.set_index('REG') 
datareg = datareg.groupby(groupby_dict, axis = 1).sum().reset_index()
```

```python
datareg
```

### Récupération du nom des régions

```python
data = pd.read_json('IRIS_population/areas_REGfr.json', typ='series')

name = []
for i in datareg['REG']:
    name.append(data[i])
# print(name)
# inclusion des noms dans une nouvelle colonne
datareg['LIBREG'] = name
```

```python
datareg[['REG','LIBREG']]
```

```python
# verification de types de données 'REG'
datareg['REG'].apply(type).unique()
```

```python
# conversion des REG 'int' en 'str'
datareg['REG'] = datareg['REG'].astype(str)
datareg['REG'].apply(type).unique()
```

```python
datareg.to_csv('IRIS_population/REG_population_France_AgeCovid.csv', index=False)
```
